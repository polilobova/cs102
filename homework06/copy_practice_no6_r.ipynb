### Задание 7

На вертикальной столбиковой диаграмме для самого популярного (popularity) и самого непопулярного жанров покажите средние значения по всем характеристикам. Отрицательные и положительные значения характеристик показывайте разными цветами.

```r
df <- music
grouped <- df %>% group_by(music_genre)  %>%
                      summarise(status = mean(popularity))
grouped
```

```r
most_popular <- (grouped %>% filter(max(status) == status))[[1]]
least_popular <- (grouped %>% filter(min(status) == status))[[1]]
ee <- subset(df, music_genre == most_popular)
yy <- subset(df, music_genre == least_popular)
```

```r
ee <- ee[,!names(ee) %in% c("instance_id", "popularity", "duration_ms")]
yy <- yy[,!names(yy) %in% c("instance_id", "popularity", "duration_ms")]
```

```r
ee <- ee[ , unlist(lapply(ee, is.numeric))] 
yy <- yy[ , unlist(lapply(yy, is.numeric))] 
```

```r
X <- as.numeric(unlist(colMeans(ee)))
Y <- names(unlist(colMeans(ee)))
```

```r
colors = list()
for(i in 1:length(X)) {                                    
  if (X[i] < 0) {
    colors[[length(colors) + 1]] <- 'azure'
  } else {
    colors[[length(colors) + 1]] <- 'pink'
  }                           
}
v = unlist(c(colors))
```

```r
par(mar=c(4,7,4,3))
barplot(X,names.arg=c(Y),
ylab="Характеристики",
col=v, 
main=most_popular, 
beside = TRUE, 
cex.names=0.5, 
ylim = c(-8, 2), 
las=1)
```

```r
X <- as.numeric(unlist(colMeans(yy)))
Y <- names(unlist(colMeans(yy)))
```

```r
colors = list()
for(i in 1:length(X)) {                                    
  if (X[i] < 0) {
    colors[[length(colors) + 1]] <- 'azure'
  } else {
    colors[[length(colors) + 1]] <- 'pink'
  }                           
}
v = unlist(c(colors))
```

```r
par(mar=c(4,7,4,1))
barplot(X,names.arg=c(Y),
ylab="Характеристики",
col=v,
main=least_popular, 
beside = TRUE, 
cex.names=0.5, 
ylim = c(-8, 2), 
#xlim = c(0, 2000), 
las=1)
```

Вывод: самый популярный по средним значениям popularity жанр - Rap, непопулярный - Anime. Для обоих единственной характеристикой, которая уходит "в минус" стала loudness. У Rap наибольшие значения принимает danceability, у Anime - energy.

### Задание 8

1. Добавьте в данные категориальный столбец `длительность_трека`, в котором будут значения: "короткая" (<=3min), "средняя" (>3min, <=5min), "длинная" (>5min). 
2. Покажите график плотности распределения данных в столбце `loudness`, сгруппированных по категориям длительности.
3. На круговой диаграмме покажите численное соотношение треков разной длительности.

```r
df <- music
```

```r
df$track_length <- df$duration_ms
df$track_length[which(df$duration_ms <= 180000)] = "короткая"
df$track_length[which(df$duration_ms > 180000 & df$duration_ms <= 300000)] = "средняя"
df$track_length[which(df$duration_ms > 300000)] = "длинная"
```

```r
song_types = list("короткая", "средняя", "длинная")
```

```r
song_types = list("короткая", "средняя", "длинная")
d <- density(subset(df, track_length == song_types[[1]])$loudness)
plot(d, col=1, ylim = c(0, 0.15), xlab = "loudness", lwd = 2, main='Плотность распределения loudness по категориям длительности')
for(i in 2:length(song_types)) {                                    
  for_type <- subset(df, track_length == song_types[[i]])
  d <- density(for_type$loudness) 
  lines(d, col = i, lwd = 2)                        
}
legend("topleft", legend=c(unlist(song_types)), col=c(1, 2, 3), lty=1:1)
```

```r
df_table <- data.frame(rbind(table(df["track_length"])))
li <- as.numeric(df_table[1, ])
n <- colnames(df_table)
```

```r
pie_labels <- paste0(n, " - ", round(100 * li/sum(li), 2), "%")
```

```r
pie(li, labels = pie_labels, main="Численное соотношение треков разной длительности")
```

Вывод: по графику плотности распределения данных в столбце loudness можно видеть, что наибольшая плотность почти достигает значения 0.15. Численное соотношение треков разной длительности на круговой диаграмме показывает, что в датасете больше всего треков средней длительности, а меньше всего длинных треков.
